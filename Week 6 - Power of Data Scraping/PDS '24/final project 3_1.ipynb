## #Handling Outlier 

#### Cek Outlier

```python
import pandas as pd


# Menghitung kuartil pertama (Q1) dan kuartil ketiga (Q3) dari setiap kolom numerik
Q1 = data.select_dtypes(include=['float', 'int']).quantile(0.25)
Q3 = data.select_dtypes(include=['float', 'int']).quantile(0.75)

# Menghitung rentang interquartile (IQR) dari setiap kolom numerik
IQR = Q3 - Q1

# Definisikan batas bawah dan batas atas untuk setiap kolom numerik
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Menghitung jumlah outlier dalam setiap kolom numerik
outlier_count = (data.select_dtypes(include=['float', 'int']) < lower_bound) | (data.select_dtypes(include=['float', 'int']) > upper_bound)

# Menampilkan jumlah outlier dalam setiap kolom numerik
outlier_count = outlier_count.sum()
outlier_count
```

```python
Q1 = data['UnitPrice'].quantile(0.25)
Q3 = data['UnitPrice'].quantile(0.75)

# Menghitung rentang interquartile (IQR) dari kolom 'UnitPrice'
IQR = Q3 - Q1

# Menentukan batas bawah dan batas atas untuk outlier dalam kolom 'UnitPrice'
lower_bound_unitprice = Q1 - 1.5 * IQR
upper_bound_unitprice = Q3 + 1.5 * IQR

# Menghitung kuartil pertama (Q1) dan kuartil ketiga (Q3) dari kolom 'Total Price'
Q1 = data['Total Price'].quantile(0.25)
Q3 = data['Total Price'].quantile(0.75)

# Menghitung rentang interquartile (IQR) dari kolom 'Total Price'
IQR = Q3 - Q1

# Menentukan batas bawah dan batas atas untuk outlier dalam kolom 'Total Price'
lower_bound_totalprice = Q1 - 1.5 * IQR
upper_bound_totalprice = Q3 + 1.5 * IQR

# Menemukan outlier dalam kolom 'UnitPrice'
outliers_unitprice = data[(data['UnitPrice'] < lower_bound_unitprice) | (data['UnitPrice'] > upper_bound_unitprice)]

# Menemukan outlier dalam kolom 'Total Price'
outliers_totalprice = data[(data['Total Price'] < lower_bound_totalprice) | (data['Total Price'] > upper_bound_totalprice)]

# Menampilkan nilai outlier dari kolom 'UnitPrice'
print("Outliers in UnitPrice:")
print(outliers_unitprice['UnitPrice'])

# Menampilkan nilai outlier dari kolom 'Total Price'
print("Outliers in Total Price:")
print(outliers_totalprice['Total Price'])
```
