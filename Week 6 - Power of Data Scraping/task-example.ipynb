```python
import requests #A library for sending HTTP requests and retrieving HTML content from websites.
from bs4 import BeautifulSoup
import bs4
import time
from urllib.request import urlopen # mengambil data dari URL dan melakukan operasi yang terkait dengan permintaan HTTP seperti mengirim permintaan GET atau POST ke server web. Fungsi urlopen() dalam modul urllib.request digunakan untuk membuka koneksi ke URL yang ditentukan dan mengembalikan objek respons.
import re
import numpy as np
import pandas as pd
```

```python
url = "https://quran.nu.or.id/al-fatihah"
req = requests.get(url)

# Memastikan permintaan berhasil
if req.status_code == 200:
    soup = BeautifulSoup(req.text)
    pretty_html = soup.prettify()
    print(pretty_html)
else:
    print(f"Failed to retrieve the webpage. Status code: {req.status_code}")
```
