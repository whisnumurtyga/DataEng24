**SCRAP ALL PRODUCT**

```python
books = []

for i in range(1,51):
    url = f"https://books.toscrape.com/catalogue/page-{i}.html"
    response = requests.get(url)
    response = response.content
    soup = BeautifulSoup(response , 'html.parser')
    ol = soup.find('ol')
    articles = ol.find_all('article', class_='product_pod')
    judul = soup.find('h1')
    for article in articles:
        image = article.find('img')
        title = image.attrs['alt']
        starTag = article.find('p')
        star = starTag['class'][1]
        price = article.find('p', class_='price_color').text
        price = float(price[1:])
        books.append([title, star, price])
        
        #print(kategori.text.replace("\n","").replace("Books","").replace(judul.text,""))
df  = pd.DataFrame(books, columns=['Title', 'Rating', 'Price'])
df
```

```python
def angka (Angka):
    if Angka == 'One' :
        return '1'
    elif Angka == 'Two' :
        return '2'
    elif Angka == 'Three' :
        return '3'
    elif Angka == 'Four' :
        return '4'
    else:
        return '5'
    
df['Rating New']=df['Rating'].apply(angka)
df
```

```python
def categori (Rating):
    if Rating == '5' :
        return 'uapik'
    elif Rating == '4' :
        return 'oke'
    elif Rating == '3' :
        return 'bolehlah'
    elif Rating == '2' :
        return 'hmmm'
    else:
        return 'gak dulu'
    
df['Review']=df['Rating New'].apply(categori)
df
```

```python
def price (harga):
    if harga >= 30.00 :
        return 'Mahal'
    elif harga >= 29.00  :
        return 'Standar'
    elif harga >= 10.00 :
        return 'Murah'
    else:
        return 'Sangat murah'
    
df['Price New']=df['Price'].apply(price)
df
```

```python
df['Euro ke Rupiah'] = df['Price'] * 17000
df
```

```python
# Memisahkan Nama berdasarkan tanda kurung dan menambahkan kolom baru
df[['Judul', 'Sub Judul']] = df['Title'].str.split('(', 1, expand=True)

# Membersihkan tanda kurung kanan pada kolom Jabatan
df['Sub Judul'] = df['Sub Judul'].str.rstrip(')')
df
```

```python
df[['Judul', 'Seri']] = df['Sub Judul'].str.split('#', 1, expand=True)

# Membersihkan spasi pada kolom Subjudul
df['Seri'] = df['Seri'].str.strip()
df
```

```python
# Mengganti nilai None dengan nilai baru ('-')
df1 = df.fillna('Tidak Ada')
df1
```

```python
def tambahkan_tanda_euro(nilai):
    return '€{:,.2f}'.format(nilai)
# Menambahkan tanda euro pada kolom 'Pendapatan (Euro)'
df1['Price'] = df1['Price'].apply(tambahkan_tanda_euro)
df1
```

```python
import requests
from bs4 import BeautifulSoup

kategori_array = []

for i in range(1, 51):
    aa = str(i)
    url1 = "https://books.toscrape.com/catalogue/page-" + aa + ".html"
    r2 = requests.get(url1)
    soup = BeautifulSoup(r2.content, 'html.parser')
    items = soup.find_all('h3')

    for item in items:
        url_main = "https://books.toscrape.com/catalogue/" + item.find('a')['href']
        r3 = requests.get(url_main)
        soup2 = BeautifulSoup(r3.content, 'html.parser')
        judul = soup2.find('h1')
        kategori = soup2.find('ul', {"class": "breadcrumb"})
        
        if kategori is not None:
            kategori_text = kategori.text.replace("\n", "").replace("HomeBooks", "")
            judul_text = judul.text if judul else ''
            kategori_array.append(kategori_text.replace(judul_text, ''))
            print(kategori_text.replace(judul_text, ''))
        else:
            print("Kategori tidak ditemukan.")
```

```python
df2  = pd.DataFrame(kategori_array, columns=['Genre'])
df2
```

```python
concat = pd.concat([df1,df2],axis=1)
concat
```

**Insight 1 : Mencari genre buku dengan rating 5 paling banyak**

```python
#menndapatkan buku dengan rating 5
filter_5 = concat[concat['Rating New']=='5']
#explode untuk memisahkan setiap genre dalam satu buku
df_genre = filter_5.explode('Genre')
#menghitung jumlah setiap genre
genre = df_genre['Genre'].value_counts()
#mencari genre yang memiliki jumlah paling banyak
genre_terbanyak = genre.idxmax()
print(f"Genre buku yang paling banyak dengan 5 rating adalah : {genre_terbanyak}")
```

**Insight 2  **

```python
#menndapatkan kolom price new 
filter_5 = concat[concat['Price New']=='Mahal']
#explode untuk memisahkan setiap genre dalam satu buku
df_genre = filter_5.explode('Genre')
#menghitung jumlah setiap genre
genre = df_genre['Genre'].value_counts()
#mencari genre yang memiliki harga paling tinggi dan rendah
genre_terbanyak = genre.idxmax()
genre_terdikit = genre.idxmin()
print(f"Genre buku dengan harga termahal adalah : {genre_terbanyak}")
print(f"Genre buku dengan harga termurah adalah : {genre_terdikit}")
```
