## PRE-PROCESSING DATA

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv(r"D:\STUDY\SEMS 3\UKM KODER\Data 2.csv")
data
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
data['Date'].fillna(data['Date'].mode()[0], inplace=True)
data['DateType'].fillna(data['DateType'].mode()[0], inplace=True)
data['Age'].fillna(data['Age'].mode()[0], inplace=True)
data['Sex'].fillna(data['Sex'].mode()[0], inplace=True)
data['DeathCity'].fillna(data['DeathCity'].mode()[0], inplace=True)

data.isnull().sum()
```

```python
data['Race'].fillna(data['Race'].mode()[0], inplace=True)
data['Location'].fillna(data['Location'].mode()[0], inplace=True)
data['InjuryPlace'].fillna(data['InjuryPlace'].mode()[0], inplace=True)
data['MannerofDeath'].fillna(data['MannerofDeath'].mode()[0], inplace=True)
data['ResidenceCityGeo'].fillna(data['ResidenceCityGeo'].mode()[0], inplace=True)
data['InjuryCityGeo'].fillna(data['InjuryCityGeo'].mode()[0], inplace=True)

data.isnull().sum()
```

```python
data['ResidenceCity'].fillna(data['ResidenceCity'].mode()[0], inplace=True)
data['ResidenceCounty'].fillna(data['ResidenceCounty'].mode()[0], inplace=True)
data['DescriptionofInjury'].fillna(data['DescriptionofInjury'].mode()[0], inplace=True)

data.isnull().sum()
```

```python
data['ResidenceState'].fillna(data['ResidenceState'].mode()[0], inplace=True)
data['LocationifOther'].fillna(data['LocationifOther'].mode()[0], inplace=True)
data['InjuryCity'].fillna(data['InjuryCity'].mode()[0], inplace=True)
data['InjuryCounty'].fillna(data['InjuryCounty'].mode()[0], inplace=True)
data['InjuryState'].fillna(data['InjuryState'].mode()[0], inplace=True)
data['OtherSignifican'].fillna(data['OtherSignifican'].mode()[0], inplace=True)

data.isnull().sum()
```

```python
# Konversi kolom Date menjadi datetime
data['Date'] = pd.to_datetime(data['Date'])

# Ekstraksi tahun
data['Year'] = data['Date'].dt.year

# Plot tren tahunan
plt.figure(figsize=(10, 6))
sns.countplot(x='Year', data=data, palette='viridis')
plt.title('Tren Tahunan Kematian')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Kematian')
plt.show()
```

```python
data['Month'] = data['Date'].dt.month

plt.figure(figsize=(10, 6))
sns.countplot(x='Month', data=data, palette='coolwarm')
plt.title('Pola Kematian Berdasarkan Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Kematian')
plt.show()
```

```python
data['DayOfWeek'] = data['Date'].dt.day_name()

plt.figure(figsize=(10, 6))
sns.countplot(x='DayOfWeek', data=data, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], palette='viridis')
plt.title('Tren Kematian Berdasarkan Hari dalam Seminggu')
plt.xlabel('Hari dalam Seminggu')
plt.ylabel('Jumlah Kematian')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Membuat scatter plot
sns.scatterplot(x='Age', y='AnyOpioid', data=data)
plt.title('Hubungan Antara Usia dan Penggunaan Opioid')
plt.xlabel('Usia')
plt.ylabel('Penggunaan Opioid')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membuat histogram
plt.figure(figsize=(12, 8))
sns.histplot(data=data, x='Race', hue='MannerofDeath', multiple='stack', palette='pink', shrink=.8)
plt.title('Jenis Kematian Berdasarkan Ras')
plt.xlabel('Ras')
plt.ylabel('Jumlah Kematian')
plt.xticks(rotation=45)
plt.show()
```

```python
# Menghitung jumlah kematian per ResidenceCounty
top_countries = data['ResidenceCounty'].value_counts().head(5).index

# Filter data berdasarkan top 5 ResidenceCounty
filtered_data = data[data['ResidenceCounty'].isin(top_countries)]

# Membuat histogram
plt.figure(figsize=(12, 8))
sns.histplot(data=filtered_data, x='ResidenceCounty', hue='MannerofDeath', multiple='stack', palette='pink', shrink=.8)
plt.title('Jenis Kematian Berdasarkan Negara (Top 5)')
plt.xlabel('Negara')
plt.ylabel('Jumlah Kematian')
plt.xticks(rotation=45)
plt.show()
```

```python
demographic_columns = ['Age', 'Sex', 'Race', 'MannerofDeath']
correlation_matrix = data[demographic_columns].apply(lambda x: pd.factorize(x)[0]).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='pink')
plt.title('Korelasi Antara Faktor Demografis')
plt.show()
```
