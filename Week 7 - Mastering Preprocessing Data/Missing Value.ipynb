```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dataset.csv', sep=";")
df
```

```python
df1 = df.loc[0:,['Name','Sex','Age']]
df1
```

```python
df1.isnull()
```

```python
df1.notna()
```

```python
df1.isnull().sum()
```

```python
df1.info()
```

```python
# Menghitung jumlah nilai "NaN" dalam kolom tertentu
count_NaN = df1['Age'].isnull().sum()
# Menghitung jumlah total data dalam kolom tertentu
total_values = df1['Age'].shape[0]
# Menghitung persentase kolom yang berisi "NaN"
percentage_NaN = (count_NaN / total_values) * 100

print(f"Persentase kolom 'Age' yang berisi nilai 'NaN' adalah: {percentage_NaN:.2f}%")
```

## DELETION

```python
# Menghapus baris yang berisi nilai 'NaN'
listwise = df1[df1 != 'NaN'].dropna()
listwise
```

```python
listwise.isnull().sum()
```

```python
listwise["Age"] = listwise["Age"].astype("int")
listwise
```

```python
listwise.info()
```

## Deletion

```python
# Menghitung nilai mean
mean = listwise['Age'].mean()

print(f"Nilai mean dari kolom 'Age' yang sudah melalui handiling missing value (deletion listwise) adalah:", mean)

# Menghitung nilai standar deviasi
std = listwise['Age'].std()

print(f"Nilai standar deviasi dari kolom 'Age' yang sudah melalui handiling missing value (deletion listwise) adalah:", std)


# Menghitung nilai varians
variance = listwise['Age'].var()

print(f"Nilai varians dari kolom 'Age' yang sudah melalui handiling missing value (deletion listwise) adalah:", variance)
```

```python
statistik = ['Mean','Standar Deviasi','Varians']
stats_des = {'Statistik Deskriptif': statistik, 'Deletion Listwise':[mean,std,variance]}
stats_des = pd.DataFrame(stats_des).set_index('Statistik Deskriptif')
stats_des 
```

## IMPUTATION

## Mengganti NaN dengan Modus

```python
# Menghitung nilai median dari kolom 'Age'
modus_age = listwise['Age'].mode()[0]
print(modus_age)
# Mengganti nilai "NaN" dengan nilai median
df1['Age'] = df1['Age'].fillna(modus_age)
df1
```

```python
# Menghitung nilai modus dari kolom 'Age'
modus_age = listwise['Age'].mode()[0]
print(modus_age)
```
