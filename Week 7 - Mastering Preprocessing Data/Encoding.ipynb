```python
from sklearn.preprocessing import LabelEncoder

# Contoh data kategori
kategori = ['merah', 'hijau', 'biru', 'merah', 'biru']

# Buat objek LabelEncoder
label_encoder = LabelEncoder()

# Terapkan Label Encoding pada data kategori
kategori_encoded = label_encoder.fit_transform(kategori)

print(kategori)
print(kategori_encoded)
```

```python
from sklearn.preprocessing import OrdinalEncoder

# Definisikan urutan kategori
urutan_kategori = ['merah', 'hijau', 'biru']
kategori = ['merah', 'hijau', 'biru', 'merah', 'biru']
kategori_array = np.array(kategori).reshape(-1, 1)

# Buat objek OrdinalEncoder dengan spesifikasi urutan
ordinal_encoder = OrdinalEncoder(categories=[urutan_kategori])

# Terapkan Ordinal Encoding pada data kategori
kategori_ordinal_encoded = ordinal_encoder.fit_transform(kategori_array)

print(kategori_array)
print(kategori_ordinal_encoded)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder

# Data kategori
kategori = ['merah', 'hijau', 'biru', 'merah', 'biru']
kategori_df = pd.DataFrame({'warna': kategori})
# Preview data kategori sebelum encoding
# print("Data kategori sebelum encoding:")
# print(kategori)

# Membuat objek OneHotEncoder
encoder = OneHotEncoder()

# Melakukan one-hot encoding dan mengubah data kategori menjadi matriks one-hot encoding
one_hot_encoding = encoder.fit_transform(np.array(kategori).reshape(-1, 1))

# Membuat DataFrame dari matriks one-hot encoding
df = pd.DataFrame(one_hot_encoding.toarray(), columns=encoder.categories_[0])
display(kategori_df, df)
```

```python
import pandas as pd

# Membuat DataFrame
data = {'Text': ['Ini adalah contoh kalimat pertama.', 'Ini kalimat kedua.'],
        'Feature1': [1, 2],
        'Feature2': [4, 3],
        'Feature3': [2, 3]}
df = pd.DataFrame(data)

# Feature Extraction
df['Word_Count'] = df['Text'].apply(lambda x: len(x.split()))
df['Average_Word_Length'] = df['Text'].apply(lambda x: sum(len(word) for word in x.split()) / len(x.split()))

# Feature Selection
selected_features = ['Feature1', 'Feature3']
df_selected = df[selected_features]

# print(df_selected)
df.head()
```
