```python
def manual_labeling(kalimat):
    # Daftar kata positif dan negatif (bisa diperluas)
    kata_positif = ["pilihan terbaik", "memberikan", "mendukung", "integritas", "peduli", "percaya", "berani", "kompeten", "responsif", "transparan", "maju", "pihak rakyat", "solusi", "pemberian", "berhasil", "bermanfaat", "meningkat", "unggul", "bagus", "mantap", 'positif']
    kata_negatif = ["ragu", "khawatir", "kurang fokus", "tidak mampu", "kecewa", "tidak yakin", "kurang kompeten", "prioritas politik", "kurang responsif", "korupsi", "kurang baik", "buruk", "mengecewakan", "lemah", "terbelakang", "rusak", "terpuruk", "kerugian", "menyedihkan", 'negatif']

    # Inisialisasi jumlah kata positif dan negatif
    count_positif = 0
    count_negatif = 0

    # Memeriksa setiap kata dalam kalimat
    for kata in kalimat.split():
        # Menghitung kata positif
        if kata in kata_positif:
            count_positif += 1
        # Menghitung kata negatif
        elif kata in kata_negatif:
            count_negatif += 1

    # Menentukan label berdasarkan jumlah kata positif dan negatif
    if count_positif > count_negatif:
        return "positif"
    elif count_negatif > count_positif:
        return "negatif"
    else:
        return "netral"
```

```python
def generate_labels(texts):
    labels = []
    for text in raw_texts:
        label = manual_labeling(text)
        labels.append(label)

    return(labels)
```

```python
labels = generate_labels(raw_texts)
print(len(labels))
print(labels)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def plot_sentimen(kalimat_sentimen, label_sentimen):
    # Misalkan Anda memiliki DataFrame seperti ini
    data = {
        'kalimat_sentimen': kalimat_sentimen,
        'label_sentimen': label_sentimen
    }

    df = pd.DataFrame(data)

    # Membuat plot
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x='label_sentimen')
    plt.title('Distribusi Sentimen')
    plt.xlabel('Sentimen')
    plt.ylabel('Jumlah')
    plt.show()
```

```python
plot_sentimen(raw_texts, labels)
```

```python
def generate_plot_word_freq(kalimat_sentimen, labels):
    # Menginisialisasi dictionary untuk menyimpan frekuensi kata-kata untuk setiap sentimen
    frekuensi_sentimen = {'positif': {}, 'negatif': {}, 'netral': {}}

    # Memisahkan kalimat berdasarkan label sentimen dan menghitung frekuensi kemunculan setiap kata
    for kalimat, label in zip(kalimat_sentimen, labels):
        kata_kata = kalimat.split()
        for kata in kata_kata:
            frekuensi_sentimen[label][kata] = frekuensi_sentimen[label].get(kata, 0) + 1

    # Menggabungkan semua grafik menjadi satu baris
    fig, axs = plt.subplots(1, 3, figsize=(18, 6))

    # Mengambil lima kata teratas untuk setiap kategori sentimen dan mengurutkannya
    for i, (sentimen, frekuensi_kata) in enumerate(frekuensi_sentimen.items()):
        # Mengurutkan frekuensi kata-kata secara menurun
        kata_terbanyak = sorted(frekuensi_kata.items(), key=lambda x: x[1], reverse=True)[:5]
        kata, frekuensi = zip(*kata_terbanyak)  # Membagi kembali menjadi dua tuple terpisah
        
        # Membuat plot batang untuk lima kata teratas
        axs[i].bar(kata, frekuensi, color='skyblue')
        axs[i].set_title(f'Lima Kata Teratas dalam Sentimen {sentimen.capitalize()}')
        axs[i].set_xlabel('Kata')
        axs[i].set_ylabel('Frekuensi')
        axs[i].tick_params(axis='x', rotation=45)  # Perbaikan disini

    plt.tight_layout()
    plt.show()
```

```python
generate_plot_word_freq(raw_texts, labels)
```

```python
def create_report(texts):
    labels = generate_labels(texts)
    plot_sentimen(texts, labels)
    generate_plot_word_freq(texts, labels)
```

```python
create_report(raw_texts)
```
